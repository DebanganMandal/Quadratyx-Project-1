```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
reviews_dataset = pd.read_csv('IPhoneReview.csv', encoding='ISO-8859-1')
```

```python
reviews_dataset.head()
```

```python
reviews_dataset.drop(columns='Unnamed: 0', inplace=True)
```

```python
reviews_dataset.info()
```

```python
sns.barplot(x=reviews_dataset['Date'].unique()[0:5], y=reviews_dataset['Date'].value_counts()[0:5])
plt.xlabel('Date')
plt.ylabel('Counts')
```

```python
reviews_dataset.drop(columns=['UserName', 'Date'], inplace=True)
```

```python
corpus = reviews_dataset['Review'].values
```

```python
list(corpus)
```

```python
Corpus = ""
for str in corpus:
    Corpus += str + " "
```

```python
Corpus
```

```python
voc = word_tokenize(Corpus)
```

```python
voc
```

```python
stop_words = list(stopwords.words('english'))
```

```python
vocubulary = []
lematizer = WordNetLemmatizer()

for word in voc:
    if (word not in stop_words and word not in string.punctuation):
        word = word.lower()
        lemmatized_word = lematizer.lemmatize(word)
        vocubulary.append(lemmatized_word)
```

```python
vocubulary
```

```python
vocubulary.remove('...')
```

```python
vocubulary.remove('..')
```

```python
vocubulary
```

```python
vocubulary = list(map(lambda x: x.replace('iphon', 'iphone'), vocubulary))
```

```python
freq = {}

for word in vocubulary:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1
```

```python
vocubulary = list(freq.keys())
```

```python
del freq['......']
del freq["''"]
```

```python
del freq['``']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()  # BOW Model of vectorizing
```

```python
X = cv.fit_transform(vocubulary)
```

```python
cv.fit(vocubulary)
```
